```python
# Data cleaning of MOVIES column:

df['MOVIES'] = df['MOVIES'].str.strip()
df.dropna(subset = ['MOVIES'],inplace = True) # drop the rows with nulls in it
df['MOVIES'] = df['MOVIES'].str.replace(r'[^A-Za-z0-9,: ]+','',regex = True)
df.drop_duplicates(subset = ['MOVIES'],inplace = True)  # drop the duplicate rows
df
```

```python
# Datacleaning for YEAR column
df['YEAR'] = df['YEAR'].str.split('–').str[0].str.strip()
df['YEAR'] = df['YEAR'].str.replace(r'[^0-9]+','',regex = True)
df = df[df['YEAR'].str.strip() != '']
df.dropna(subset = ['YEAR'],inplace = True)

df['YEAR'] = df['YEAR'].astype(int)

df
```

```python
df.info()
```

```python
# To replace all the newline \n with blanks:
df['GENRE'] = df['GENRE'].str.replace('\n','',regex = False)
df
```

```python
# To fill the blanks in ratings by average of each genre:
df['GENRE'] = df['GENRE'].str.strip()
df['Rating missing'] = df['RATING'].isnull() # created a new column to keep the flag available where the data is originally missing for Ratings column
df
```

```python
df['RATING'] = df.groupby('GENRE')['RATING'].transform(lambda x:x.fillna(x.mean()))  #average values filling in blank rows
df['RATING'] = df['RATING'].round(2) #Round off function
df
```

```python
# Replacing newlines, tabs, carriage returns with blanks

df['ONE-LINE'] = df['ONE-LINE'].str.strip()
df['ONE-LINE'] = df['ONE-LINE'].str.replace(r'[\n\r\t]+',' ',regex = True)
df['ONE-LINE'] = df['ONE-LINE'].str.replace(r'\s+',' ',regex = True) #replace multiple spaces with single space
df['ONE-LINE'] = df['ONE-LINE'].fillna('No Description Available',inplace = False) #replace empty rows with a description
df
```

```python
df['STARS'] = df['STARS'].str.replace('\n',' ',regex = False)
df['STARS'] = df['STARS'].str.replace(r'\s+',' ',regex = True)
df['STARS'] = df['STARS'].str.strip()
df
```

```python
df[['DIRECTOR_RAW','STARS_RAW']] = df['STARS'].str.split('|',n=1,expand = True) # we have splitted into director and stars column
df
```

```python
df['DIRECTOR'] = df['DIRECTOR_RAW'].str.replace('Director: ','',regex = False)
df['STARS'] = df['STARS_RAW'].str.replace('Stars: ','',regex=False)
df
```

```python
#few of the stars are at director column due to director names are not present.To avoid this follow this process:
mask = df['DIRECTOR'].str.startswith('Stars: ',na=False) & df['STARS'].isna()
df.loc[mask,'STARS'] = df.loc[mask,'DIRECTOR'].str.replace('Stars: ','',regex = False).str.strip(' ,')
df.loc[mask,'DIRECTOR'] = None
df['DIRECTOR'] = df['DIRECTOR'].str.replace('Directors: ','',regex=False)
df = df.drop(columns = ['DIRECTOR_RAW','STARS_RAW'])
df
```

```python
df['VOTES'] = df['VOTES'].fillna(0)
#print(df['VOTES'].dtype)
#df[df['VOTES'].astype(str).str.contains(r'\s')]
df
```

```python
df['RunTime'] = df.groupby('GENRE')['RunTime'].transform(lambda x:x.fillna(x.mean()))
df['RunTime'] = df['RunTime'].round(2)
df
```
